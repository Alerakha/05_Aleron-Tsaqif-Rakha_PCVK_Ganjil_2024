```python
import numpy as np
import cv2 as cv
from PIL import Image
from google.colab.patches import cv2_imshow as cv_imshow
import matplotlib.pylab as plt
import matplotlib.patches as patches

img_pil = Image.open('KTP.jpg')
img = np.array(img_pil)
img = cv.resize(img, (0,0), fx=0.5, fy=0.5)
tinggi = img.shape[0]
lebar = img.shape[1]
fig, ax = plt.subplots()
# image_bgr = cv.rectangle(img.copy(), (100, int(tinggi/2)), (250, int(tinggi/2)), (255, 255, 255), 2)
image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# img = cv.rectangle(img, (513,165), (535,185), (255,255,255), -1)
# img = cv.rectangle(img, (600,385), (765,480), (255,255,255), -1)

img = cv.rectangle(img, (260,84), (282,93), (255,255,255), -1)
img = cv.rectangle(img, (300,193), (382,240), (255,255,255), -1)
final_frame = cv.hconcat((img, image_rgb))

rect1 = patches.Rectangle((513, 165), 40, 25, linewidth=1, edgecolor='black', facecolor='white')
rect2 = patches.Rectangle((600, 385), 170, 85, linewidth=3, edgecolor='black', facecolor='yellow')



ax = plt.gca()
ax.add_patch(rect1)
ax.add_patch(rect2)
cv_imshow(final_frame)
plt.imshow(img_pil)
plt.show()

# cv.imshow('image', img)
# cv.waitKey(0)
# cv.destroyAllWindows()
```
