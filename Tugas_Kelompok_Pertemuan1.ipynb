2. Matplotlib tidak memerlukan konversi dari BGR ke RGB, karena warna default Matplotlib untuk memproses gambar adalah RGB. Sedangkan opencv perlu menggunakan konversi BGR ke RGB dengan ``` cv.COLOR_BGR2RGB ``` karena format warna default opencv untuk memproses gambar adalah BGR

````python
import numpy as np
import cv2 as cv
from PIL import Image
from google.colab.patches import cv2_imshow as cv_imshow
import matplotlib.pylab as plt
import matplotlib.patches as patches

img_pil = Image.open('KTP.jpg')
img = np.array(img_pil)
img = cv.resize(img, (0,0), fx=0.5, fy=0.5)
# fig, ax = plt.subplots()
image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

print('CV2 mengubah langsung format ke RGB')
img2 = cv.imread('KTP.jpg')
cv_imshow(img2)
# print('Jika ditambahkan resize, auto format RGB tidak dibaca')
# img2 = cv.imread('KTP.jpg')
# img2 = cv.resize(img, (500,300))
# cv_imshow(img2)

final_frame = cv.hconcat((img, image_rgb))
print("Matplotlib tanpa BGR to RGB")
plt.imshow(img)
plt.show()
# Tugas
rect1 = patches.Rectangle((190, 135), 300, 25, linewidth=1, edgecolor='black', facecolor='white')
rect2 = patches.Rectangle((190, 163), 170, 22, linewidth=1, edgecolor='black', facecolor='yellow')

ax = plt.gca()
ax.add_patch(rect1)
ax.add_patch(rect2)
print("OpenCV: kiri tanpa BGR to RGB, kanan dengan BGR to RGB")
cv_imshow(final_frame)
print("Tugas Kelompok 2 :")
plt.imshow(img_pil)
# plt.axis('off')
plt.show()
````

- Dibandingkan dengan OpenCV yang harus menginclude kan 'img' dalam fungsi rectangle nya. Matplotlib akan lebih fleksibel dalam menampilkan objek dalam gambar, karena objek disimpan dalam variabel.

- Matplotlib khusus digunakan untuk menampilkan visual perhitungan grafik, persebaran data, histogram dan visualisasi perhitungan plot lainnya.
