```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')

# image = cv2.imread('/content/drive/MyDrive/PCVK/Images/KTP.jpeg')
image = cv2.imread('KTP.jpg')
mask = np.zeros(image.shape, dtype=np.uint8)
mask = cv2.rectangle(mask, (10,280), (400,300), (255, 255, 255), -1)
mask = cv2.rectangle(mask, (10,305), (400,325), (255, 255, 255), -1)

result_and = cv2.bitwise_and(image, mask)
result_and[mask==0] = 255 #untuk menghilangkan warna hitam pada mask untuk memperjelas seleksi

# final_frame = cv2.hconcat([image, mask,result_and])
# cv2_imshow(final_frame)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
result_and = cv2.cvtColor(result_and, cv2.COLOR_BGR2RGB)

plt.subplot(2,2,1)
plt.title('Gambar Awal')
plt.axis('off')
plt.imshow(image)

plt.subplot(2,2,2)
plt.title('Mask')
plt.axis('off')
plt.imshow(mask)

plt.subplot(2,2,3)
plt.title('Hasil seleksi dengan AND')
plt.axis('off')
plt.imshow(result_and)
```
